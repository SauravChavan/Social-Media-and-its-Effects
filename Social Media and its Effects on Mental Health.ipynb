### Import the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px

%matplotlib inline
pd.set_option('display.max_columns', 150)
```

```python
data = pd.read_excel("data.xlsx")
data.drop(columns=['Timestamp','Email Address'], inplace=True)
data.rename(columns = {'Enter name':'Name', 'Are you a college student?':'In College?', 
                       'Are you a working professional?':'Working?', 'Please enter college/work city':'City', 
                       'Enter your app preference [Instagram]':'Instagram', 'Enter your app preference [Twitter]':'Twitter',
                       'Enter your app preference [Facebook]':'Facebook', 'Enter your app preference [Other]':'Other',
                       'Types of pages followed:':'Pages followed', 
                       'Enter number of followers/online friends (as close to original as possible)':'Followers',
                       'Do you use social media for professional purposes also(work, academics, connecting with new people, etc)?':'Professional Use?',
                       'Number of hours spent on social media daily (approx)':'Hours spent', 
                       'Rate your mental health on a scale of 1-5':'Mental Health',
                       "Do you feel excess usage of social media takes a toll on one's mental health?":"Excess use has bad effects?",
                       "Do you feel there's more hate on social media than positivity and that it affects one's mood?":'SM is +ve/-ve?',
                       'Do you feel we are more dependent on social media due to this pandemic than before?':'Increase in time spent on SM after Lockdown',
                       'Mark the option(s) which you feel have a correlation with excess usage of social media':'Effects of excess use',
                       'Do you feel that usage of social media has any of the following benefits on mental health during the pandemic?':'Benefits',
                       "Do you feel that social media does more harm than benefit an individual's mental health?":"More harm or more benefit?",
                       'Do you find yourself using social media more in leisure time?':'More use in leisure time?',
                       "Do you wonder you could spend your time in a more productive way if it wasn't for social media?":'Time is more productive without SM?'},
                       inplace=True)
data.index += 1
data = data.fillna('None')
data.head(5)
```

```python
for l in range(1,data.shape[0]+1):
    data['Name'][l] = data['Name'][l].lower()
    data['Name'][l] = data['Name'][l].title()
```

```python
data.loc[data['In College?']=='Yes', 'In College?'] = 1
data.loc[data['In College?']=='No', 'In College?'] = 0

data.loc[data['Working?']=='Yes', 'Working?'] = 1
data.loc[data['Working?']=='No', 'Working?'] = 0

data.loc[data['Professional Use?']=='Yes', 'Professional Use?'] = 1
data.loc[data['Professional Use?']=='No', 'Professional Use?'] = 0

data.loc[data['Excess use has bad effects?']=='Yes', 'Excess use has bad effects?'] = 1
data.loc[data['Excess use has bad effects?']=='No', 'Excess use has bad effects?'] = 0
data.loc[data['Excess use has bad effects?']=='Maybe', 'Excess use has bad effects?'] = -1

data.loc[data['SM is +ve/-ve?']=='Yes', 'SM is +ve/-ve?'] = 1
data.loc[data['SM is +ve/-ve?']=='No', 'SM is +ve/-ve?'] = 0
data.loc[data['SM is +ve/-ve?']=='Maybe', 'SM is +ve/-ve?'] = -1

data.loc[data['More harm or more benefit?']=='Yes', 'More harm or more benefit?'] = 1
data.loc[data['More harm or more benefit?']=='No', 'More harm or more benefit?'] = 0
data.loc[data['More harm or more benefit?']=='Maybe','More harm or more benefit?'] = -1

data.loc[data['More use in leisure time?']=='Yes', 'More use in leisure time?'] = 1
data.loc[data['More use in leisure time?']=='No', 'More use in leisure time?'] = 0
```
